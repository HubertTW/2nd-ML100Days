```python
# 載入基礎套件
import numpy as np
np.random.seed(1)

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#負相關

x = np.random.randint(0, 50, 1000)


y = -x+np.random.normal(0, 10, 1000)

np.corrcoef(x, y)
```

```python
# 將分布畫出來看看吧
plt.scatter(x, y)
```
